```python
import pandas as pd

df = pd.read_csv(r"C:\Users\aayus\Downloads\student-mat.csv", sep=";")
print(df.head())
print(df.columns)
```

Step 4: Calculate the Average Final Grade

The average value of the final grade (`G3`) is calculated to understand the overall academic performance of students

```python
print(df["G3"].mean())
```

Step 5: Students Scoring Above 15

This analysis counts the number of students who scored more than 15 marks in the final examination

```python
above_15 = df[df["G3"] > 15]
print("Number of students scoring above 15:", len(above_15))
```

Step 6: Correlation between Study Time and Performance

The correlation between study time and final grades is calculated to determine whether increased study time is associated with better academic performance

```python
correlation = df["studytime"].corr(df["G3"])
print("Correlation between study time and Final Grade:", correlation)
```

Step 7: Average Performance by Gender

The average final grades of male and female students are compared to identify which group performs better on average

```python
gender_avg = df.groupby("sex")["G3"].mean()

plt.figure(figsize=(6,5))
plt.bar(gender_avg.index, gender_avg.values)
plt.title("Average Final Grade by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Final Grade (G3)")
plt.show()
```

Step 8: Histogram of Final Grades

A histogram is used to visualize the distribution of students' final grades. It helps identify the most common score range

```python
plt.figure(figsize=(8,5))
plt.hist(df["G3"], bins = 10, edgecolor = "black")
plt.title("Histogram of final grades(G3)")
plt.xlabel("Final Grade(G3)")
plt.ylabel("Number of students")
plt.grid(True)
plt.show()
```

Step 9: Scatter Plot of Study Time vs Final Grade

A scatter plot is used to observe the relationship between study time and final grades. Each point represents one student's data

```python
plt.figure(figsize=(8,5))
plt.scatter(df["studytime"], df["G3"])
plt.title("Study Time vs Final Grade")
plt.xlabel("Study Time")
plt.ylabel("Final Grade (G3)")
plt.grid(True)
plt.show()
```

Step 10: Gender-wise Average Final Grade

A bar chart is created to compare the average final grades of male and female students

```python
gender_avg = df.groupby("sex")["G3"].mean()

plt.figure(figsize=(6,5))
plt.bar(gender_avg.index, gender_avg.values)
plt.title("Average Final Grade by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Final Grade (G3)")
plt.show()
```
